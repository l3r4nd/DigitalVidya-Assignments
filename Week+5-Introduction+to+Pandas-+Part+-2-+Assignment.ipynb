```python
#10. Add a column 'run_run' which checks total_runs column and say 
#'Duck' if total_runs is 0, 
#'come_on' if total_runs is 1 or 2,
# 'cool' if total_runs is 3,
# 'hit!' if total_runs is 4 or 6. 
# Hint: you can write a function using nested if-else and apply that function to df.total_runs
def run_run(number):
    if number == 0:
        return 'Duck'
    elif number == 1 or number == 2:
        return 'come_on'
    elif number == 3:
        return 'cool'
    elif number == 4 or number == 6:
        return 'hit!'
    else :
        return '_'

deliveries['run_run'] = deliveries['total_runs'].apply(lambda x: run_run(x))
```

```python
deliveries['run_run1'] = deliveries['total_runs'].apply(lambda x: 'Duck' if x == 0 else ('come_on' if x == 1 or x == 2
                                                                                   else ('cool' if x == 3 
                                                                                   else ('hit!' if x == 4 or x == 6
                                                                                   else '_'))))
```
